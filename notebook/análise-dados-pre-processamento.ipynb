```python
#Importação das bibliotecas utilizadas

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler, PowerTransformer
```

```python
#Abrindo os dados utilizados

df_1 = pd.read_csv('PIB-municipios.csv')
df_2 = pd.read_csv('CAR-recursos-naturais-municipios.csv')
df_3 = pd.read_csv('IDH-cidades.csv')
df_4 = pd.read_csv('IBGE-municipios.csv')
```

```python
#Juntando todos os dados em uma única variável

df = pd.merge(df_1, df_2, on="Nome do Município")
df = pd.merge(df, df_3, how='inner', on="Nome do Município")
df = pd.merge(df_2, df_3, how='inner', on="Nome do Município")
df = pd.merge(df, df_4, how='inner', on="Nome do Município")

#Removendo os atributos categóricos
df = df.select_dtypes(include=np.number)

#len(df)
```

```python
#Verificação visual de outliers

#Entender esse código

def boxplots(df, ncolunas):
  natributos = len(df.columns)
  nlinhas = natributos // ncolunas + (natributos % ncolunas != 0)
  fig, axes = plt.subplots(nlinhas, ncolunas, figsize=(5*ncolunas, 5*nlinhas))
  for x,atributo in enumerate(df.columns):
    i = x // ncolunas
    j = x % ncolunas
    sns.boxplot(data=df, x=atributo, ax=axes[i,j])
    axes[i,j].grid()

  for x in range(j+1, ncolunas):
    fig.delaxes(axes[i,x])
  plt.show()

boxplots(df, 5)
```

```python
#Entender esse código

# Remoção dos outliers (usando IIQ)
a = 1.75

discrepantes = set() # conjunto vazio

for atributo in df.columns:
  x = df[atributo]
  q = np.quantile(x, [0.25, 0.5, 0.75, 1.0])
  iiq = q[2] - q[0]
  indices = x[(x < (q[0] - a*iiq)) | (x > (q[2] + a*iiq))].index
  indices = set(indices)
  discrepantes = discrepantes.union(indices)

discrepantes = list(discrepantes)

outliers = df.iloc[discrepantes]
tipicos = df.drop(discrepantes)

noutliers = len(outliers)
ntipicos = len(tipicos)
ntotal = len(df)

poutliers = noutliers / ntotal * 100
ptipicos = ntipicos / ntotal * 100

stats = (noutliers, ntotal, poutliers, ntipicos, ptipicos)
print('%d de %d registros discrepantes (%.3f%%) e %d registros comuns (%.3f%%)' % stats)
```

```python
#Normalização dos valores

#Entender esse código

def distribuicoes(df, ncolunas):
  natributos = len(df.columns)
  nlinhas = natributos // ncolunas + (natributos % ncolunas != 0)
  fig, axes = plt.subplots(nlinhas, ncolunas, figsize=(5*ncolunas, 5*nlinhas))
  for x,atributo in enumerate(df.columns):
    i = x // ncolunas
    j = x % ncolunas
    sns.kdeplot(data=df, x=atributo, ax=axes[i,j])
    axes[i,j].grid()

  for x in range(j+1, ncolunas):
    fig.delaxes(axes[i,x])
  plt.show()

distribuicoes(df, ncolunas=5)
```
